**¿Qué es una neurona?**

Es un nombre elegante que le pusieron para referirse a una función en los años de 1940-1950. Se creeía que los nodos eran como neuronas del cerebro.  

**¿Qué es una función?**

Es un nombre elegante que pusieron para referirse a algo que toma cosas de entrada, aplica alguna lógica y saca un resultado. 

**¿Qué es una red neuronal?**

Es la interconexión de múltiples neuronas (funciones) para resolver un problema complejo que una sola neurona no podría. Dentro de los componentes existen *nodos* y *conexiones*

**Pasos que hace la red neuronal:**

1. Combinación de las variables de entrada (X). Cada señal recibida tiene una constante w (peso/weight) que permite manipular la importancia de cada variable de entrada. 
2. Neurona de respuesta. Considerando todas las variables de entrada, la neurona genera una señal de respuesta a través de una función. 

**Tipos de capas en la red neuronal**

- Capa de entrada: capa que toma los datos de entrada
- Capas ocultas: toman las entradas de otra capa y pasan a salida hacia otra capa (son intermediarias entre la entrada y la salida)
- Capas de salida: capa que hace la predicción

**Funciones de activación**

El comportamiento de las neuronas de las redes neuronales es basado en la función en la que la neurona se especializó. Una neurona puede cambiar su respuesta dependiendo de la función de activación $\varphi(*)$ que sea seleccionada

Funciones comunes:
1. Lineal: $\varphi(\nu)=\nu$
2. Sigmoidal: $\varphi(\nu)=\frac{e^{\nu}}{1+e^{\nu}}$ 
3. Tanh: $\varphi(\nu)=\frac{e^{\nu}-e^{-\nu}}{e^{\nu}+e^{-\nu}}=tanh(\nu)$
4. Softplus: $\varphi(\nu)=log(1+e^{\nu})$
5. RELU: $\varphi(\nu)=max(0,\nu)$

Las funciones de activación suelen ser diferenciables, lo que significa que la derivada de primer orden se puede calcular para un valor de entrada dado. Esto es necesario dado que las redes neuronales generalmente se entrenan utilizando el algoritmo de backpropagation que requiere la derivada del error de predicción para actualizar los pesos del modelo.

**¿Cómo elegir la función de activación para las capas ocultas?**

- Lineal: se recomienda para cuando el problema es sencillo
- Sigmoidal: típicamente para problemas de clasificación. Sufre del problema de "Gradiente que se desvanece", los gradientes de la salida de la red con respecto a los parámetros se vuelven tan pequeños que la red neuronal da pasos más pequeños hacia los mínimos de la función de costo y finalmente deja de aprender. 
- Tanh: muy similar a la sigmoidal
- RELU: es la más utilizada para capas ocultas. Es simple de implementar y no es tan susceptible a gradientes que se desvanecen
- Softmax: no se recomienda mucho usarla ya que el cálculo de gradientes no es tan eficiente

**¿Cómo son las redes neuronales para regresión?**

$$\nu^{1} = w_{0}^{1}+w^{1}$$
$$y^{1} = \varphi(\nu^{1})$$
$$\nu^{2} = w_{0}^{2}+w^{2}y_{1}$$
<font color= #2E9AFE>$$y^{2} = \varphi(\nu^{2})$$</font>

La única diferencia es la última salida. La función de la salida tiene que ser **lineal**

**Backpropagation**

Es el método que usa para ajustar el modelo. Minimiza el error medio cuadrático (MSE), utilizando el gradiente descendente para calcular los parámetros con el orden del último hacia el primero.

**¿Cómo elegir cuántas capas ocultas?**

Se recomienda usar la regla de:
(# capas ocultas = #entradas + # salidas)

**¿Cómo elegir cuántas neuronas en las capas ocultas?**

Depende....

De forma ideal, el número de neuronas ocultas debe ser mayor al número de entradas (#Neuronas ocultas > # variables de entrada). 

Peroooo esta regla tiene varias formas de verse afectada:

1. Número de datos
2. Poder computacional

Si se tienen pocos datos, al aumentar el número de neuronas ocultas, la cantidad de pesos se empieza a ser comparable con las muestras que se tienen, entonces conviene aumentar la cantidad de neuronas ocultas. 

Si se tienen muchos datos, esta regla de (#Neuronas ocultas > # variables de entrada) no es fácil por cuestiones computacionales, por lo tanto usamos las siguientes reglas:

- El número de neuronas ocultas debe estar entre el tamaño de la capa de entrada y el tamaño de la capa de salida.
- El número de neuronas ocultas debe ser 2/3 del tamaño de la capa de entrada, más el tamaño de la capa de salida.
- El número de neuronas ocultas es la media de las neuronas en las capas de entrada y salida.

**Recomendaciones**

- Se recomienda hacer un pre-procesamiento de los datos - Escalamiento
- Comúnmente todas las capas ocultas usan la misma función de activación
- La función de salida generalmente usa diferente función de activación que el de las capas ocultas. 
- En la experiencia, conviene más aumentar el número de neuronas ocultas en lugar de aumentar el número de capas ocultas.
- Se recomienda no usar muchas capas ocultas


La funcion de activacion debe de ser lineal para la capa de la salida en la regresion

#### Ejemplo de una sola variable (X) 

```python
#Generamos datos
#Generar la X
m = 1000;
X = np.linspace(-1,1,num=m)
X = np.reshape(X,(m,1))

#Generar la Y

#Y = X**2 #Función cuadrática
Y = 2*X**3+2*X**2 
#Y = 4*X*np.sin(10*X)+X*np.exp(X)*np.cos(20*X) 

[m,ne] = np.shape(X)

from sklearn import preprocessing
X = preprocessing.scale(X)
Y = preprocessing.scale(Y)

#graficando los datos
fig = plt.figure()
plt.plot(X,Y, linewidth=3)
plt.grid()
```

```python
#Dividimos los datos en test y train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.2, random_state=42)
```

#### capas son  cada etapa del modele y la capas ocultas ocurren dentro de este

```python
# Diseño del modelo

#Inicializo objeto
model = Sequential()

#Agregar capas

#Agregamos la primera capa de entrada
#capa densa significa que todas las neuronas anteriores van con todas las que siguen
#Elegimos 5 neuronas en esa capa
#input_dim = 1 porque tenemos solo una X. Aquí nos preguntamos: ¿cuántas entradas voy a tener?
model.add(Dense(5, activation='relu', input_dim=1,name='input_layer'))

#Dense es denso es todo con todo y le pido 5 neuronas, la funcion de activacion, datos de entrada y como se va a llamar la capa

#si quisiéramos añadir capas ocultas
#model.add(Dense(10, activation='tanh', name='hidden_layer'))

#Agregamos la capa de salida
###la capa de salida tiene que ser lineal
model.add(Dense(1, activation='linear',name='output_layer'))

## Lo de arriba apenas es la estructura


#gradiente descendente: brinca hasta encontrar el valor optimo
#épocas = cuántas veces se repite el gradiente descendente para encontrar el mínimo del MSE
#si la funcion de costo J tiene muchos picos, se recomienda disminuir el decaimineto o aumentar la tasa de aprendizaje
#Momentum: Utiliza la ley física del movimiento para atravesar óptimos locales (pequeñas colinas). Ayuda a que no de tantos saltos y llegue mas rapido al hoptimo
#agrega un impulso que hará que la convergencia sea más rápida

#Hiperparámetros: 1

learning_rate=0.1 #Tasa de aprendisaje o tmaño del paso
epochs = 200 # #épocas = cuántas veces se repite el gradiente descendente para encontrar el mínimo del MSE
momentum = 0.8 #Encuentra un valor optimo para que no de tantos brincos el algoritmo
decay_rate = learning_rate/epochs #Numero de decaimiento
sgd = SGD(lr=learning_rate, decay=decay_rate, momentum=momentum) #objeto del gradiente desendiente


model.compile(loss='mean_squared_error', #fucnion de costo
              optimizer=sgd, #metrica a utilizar
             metrics = ['mse']) #optimizador 
```

```python
#Ajustamos el modelo
history = model.fit(X_train, y_train, 
                   epochs=epochs, #Son hiperparametros de la red Toma el proceso de mini baches y hace muchas iteraciones(lo hace muchas veces cambiando datos)
                   batch_size=100, #Divide en baches el gradiente desendiente para encontrar el tamaño optimo de los pesos
                   validation_data=(X_test,y_test))
```

```python
#Graficar el MSE minetras mas pequeño mejor
fig, ax = plt.subplots(1, 1, figsize=(10,6))
ax.plot(np.sqrt(history.history['mse']), 'r', label='train')
ax.plot(np.sqrt(history.history['val_mse']), 'b' ,label='val')
ax.set_xlabel(r'Epoch', fontsize=20)
ax.set_ylabel(r'MSE', fontsize=20)
ax.legend()
ax.tick_params(labelsize=20)
#Si haay muchos picos hay mucho error
```

```python
#Evaluación del modelo

#Evaluación del modelo en los datos de prueba
score = model.evaluate(X_test,y_test)
print('Test mse:', score[1])
```

```python
#Predicción  y calcular R2

#Con datos del test
yhat = model.predict(X_test)
R2_score = r2_score(y_test,yhat)
xmin,xmax = min(y_test),max(y_test)
xline = np.linspace(xmin,xmax)
fig = plt.figure(figsize=(10,6))
plt.scatter(y_test,yhat,label='Estimation')
plt.plot(xline,xline,'k--',label='Perfect estimation')
plt.xlabel('Real output', fontsize=20)
plt.ylabel('Estimation output', fontsize=20)
plt.title('Test R^2=%0.4f'%R2_score, fontsize=20)
plt.legend()
plt.grid()
plt.show()
```

```python
print('R2:', R2_score)
```

#### Ejemplo Cross Validation

```python
### Cross Validation
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
```

```python
# Crear función que regresa una red neuronal compilada
def create_network():
    
    # Iniciar objeto de red neuronal
    network = Sequential()

    # Añadir una capa de entrada
    network.add(Dense(5, activation='tanh', input_dim=2,name='input_layer')) #Relu

    # Añadir capa oculta
    #network.add(Dense(10, activation='relu')) #tanh

    # Añadir capa de salida
    network.add(Dense(1, activation='linear'))
    
    opt = keras.optimizers.SGD(lr=learning_rate,momentum=momentum,
                           decay=decay_rate)
    
    # Compilar la red neuronal
    network.compile(loss='mean_squared_error', 
                    optimizer=opt, 
                    metrics=['mse']) 
    
    #regresa la red compilada
    return network
```

```python
# Hacer el modelo en keras para que pueda ser usado por scikit-learn
neural_network = KerasRegressor(build_fn=create_network, #Hacer la red con mcuchos mini test, train split
                                 epochs=200, 
                                 batch_size=100, 
                                 verbose=0)
```

```python
# Evaluar la red neuronal usando 10-fold cross-validation
print('MSE:', (-cross_val_score(neural_network, X, Y, cv=10,  scoring='neg_mean_squared_error')).mean())
```

```python
print('R2:', (cross_val_score(neural_network, X, Y, cv=10,  scoring='r2')).mean())
```

### Optimización de hiperparámetros

**GridSearch**

![GridSearch](https://www.yourdatateacher.com/wp-content/uploads/2021/03/image-7.png)

El gridsearch toma los valores de la tasa de aprendisaje crea una convinacion como una  maya con todos los valores que puede tener un hiperparametro, calcula un modelo y el la medida de performance y con la mejor convinacion metes esos valores de hiperprametros. (es con crosvalidation)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score,mean_squared_error,make_scorer
```

```python
#Crear la funcion con la estructura de la red
def create_model(lr=0.1,momentum=0.8):
    # Neural network architecture
    model = Sequential()
    model.add(Dense(5,activation='tanh',input_dim=2))
    model.add(Dense(1,activation='linear'))
    
    # Optimizer configuration
    opt = keras.optimizers.SGD(lr=lr,momentum=momentum)
    model.compile(loss = 'mean_squared_error',
                  optimizer=opt,
                  metrics=['mse'])
    return model

epochs = 200
batch_size = 100
#learning_rate = 0.1
#decay_rate = learning_rate/epochs
#momentum = 0.8

model_search = KerasRegressor(build_fn=create_model,epochs=epochs)

# Definir los parámetros del grid search
lr = [0.1, 0.05, 0.01] #busca de estos 3 valores caul es dl mejor
momentum = [0.8, 0.6, 0.4] #igual aqui
param_grid = dict(lr=lr,momentum=momentum)#malla

selection_score = make_scorer(r2_score) #r2
#selection_score = make_scorer(mean_squared_error,greater_is_better=False) #MSE

grid = GridSearchCV(estimator=model_search, param_grid=param_grid,
                    cv=2,return_train_score=True,
                    scoring=selection_score) # meter modelo, maya, cuantos falls, ver cual es la medida de performance y cual va a ser la medida

grid_result = grid.fit(X_train, y_train) #Entrenar la maya

# Imprimir resultados
# means es de la R2 que pusimos antes
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print(f' mean={mean:.4}, std={stdev:.4} using {param}')
```

```python
#Es un paso intermedio en la parte de construir la red
print('Mejores parámetros:')
print(grid_result.best_params_) #Me da los mejres parametros y el mejor score
print('Mejores score:')
print(grid_result.best_score_)
my_model= grid_result.best_estimator_
#my_model.model.summary()
weights = my_model.model.get_weights() #pesos
```

### Ventajas y desventajas de las redes neuronales

#### Ventajas

- Capacidad de trabajar con información incompleta
- Si algo llega a fallar en una neurona, se pueden aún obtener resultados - Tolerantes a fallas
- Tienen la capacidad de aprender por sí mismas y generar resultados que no se limitan a la información que se les dió al principio.
- Capacidad de procesamiento en paralelo
- Autoorganización: la capacidad de agrupar y clasificar grandes cantidades de datos hace que las redes neuronales sean especialmente adecuadas para organizar los complicados problemas visuales que plantea el análisis de imágenes

#### Desventajas

- Debido a su naturaleza de "caja negra" se pierde interpretación del modelo
- Usualmente requieren muchos más datos que los modelos tradicionales
- Si hay demasiados datos, la red neuronal requiere procesadores con gran poder de procesamiento
- No hay regla específica para determinar la estructura de la red - Prueba y error
- Pueden llegar a convertirse muy complejas
